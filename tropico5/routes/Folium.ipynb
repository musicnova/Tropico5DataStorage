```python
map_1 = folium.Map(location=WHERE, zoom_start=9, tiles='Stamen Terrain')
folium.Marker(FIRST, popup='Mt. Hood Meadows').add_to(map_1)
folium.Marker(SECOND, popup='Timberline Lodge').add_to(map_1)
map_1
```

```python
map_1 = folium.Map(location=WHERE,
                   zoom_start=9,
                   tiles='Stamen Terrain')
folium.Marker(FIRST,
              popup='Mt. Hood Meadows',
              icon=folium.Icon(icon='cloud')
             ).add_to(map_1)
folium.Marker(SECOND,
              popup='Timberline Lodge',
              icon=folium.Icon(color='green')
             ).add_to(map_1)
folium.Marker(THIRD,
              popup='Some Other Location',
              icon=folium.Icon(color='red',icon='info-sign')
              ).add_to(map_1)
map_1
```

```python
map_2 = folium.Map(location=WHERE,
                   tiles='Stamen Toner',
                   zoom_start=9)
folium.Marker(FIRST,
              popup='The Waterfront'
             ).add_to(map_2)
folium.CircleMarker(SECOND,
                    radius=5,
                    popup='Laurelhurst Park',
                    color='#3186cc',
                    fill_color='#3186cc',
                   ).add_to(map_2)
map_2
```

```python
map_3 = folium.Map(location=WHERE, tiles='Stamen Terrain', zoom_start=9)
map_3.add_child(folium.LatLngPopup())
map_3
```

```python
map_4 = folium.Map(location=WHERE,
                   tiles='Stamen Terrain',
                   zoom_start=9)
folium.Marker(FIRST, popup='Camp Muir').add_to(map_4)
map_4.add_child(folium.ClickForMarker(popup="Waypoint"))
map_4
```

```python
map_5 = folium.Map(location=WHERE, zoom_start=9)

folium.RegularPolygonMarker(
    FIRST,
    popup='Ross Island Bridge',
    fill_color='#132b5e',
    number_of_sides=3,
    radius=10
    ).add_to(map_5)
folium.RegularPolygonMarker(
    SECOND,
    popup='Hawthorne Bridge',
    fill_color='#45647d',
    number_of_sides=4,
    radius=10
    ).add_to(map_5)
folium.RegularPolygonMarker(
    THIRD,
    popup='Steel Bridge',
    fill_color='#769d96',
    number_of_sides=6,
    radius=10
    ).add_to(map_5)
folium.RegularPolygonMarker(
    FORTH,
    popup='Broadway Bridge',
    fill_color='#769d96',
    number_of_sides=8,
    radius=10
    ).add_to(map_5)
map_5
```

```python
import json

buoy_map = folium.Map(
    FIRST,
    zoom_start=7,
    tiles='Stamen Terrain'
    )

folium.RegularPolygonMarker(
    SECOND,
    fill_color='#43d9de',
    radius=12,
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open(WORKDIR+'/vis1.json')), width=450, height=250))
    ).add_to(buoy_map)

folium.RegularPolygonMarker(
    THIRD,
    fill_color='#43d9de',
    radius=12,
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open(WORKDIR+'/vis2.json')), width=450, height=250))
    ).add_to(buoy_map)

folium.RegularPolygonMarker(
    FORTH,
    fill_color='#43d9de',
    radius=12,
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(json.load(open(WORKDIR+'/vis3.json')), width=450, height=250))
    ).add_to(buoy_map)

buoy_map
```
