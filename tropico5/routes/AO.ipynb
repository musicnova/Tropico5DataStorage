```python
import os
data_dir = '/home/user/ALBINA/GALINA/KAMERA'

import pandas as pd
df = pd.read_csv(os.path.join(data_dir, 'result.csv'), sep=';', encoding='utf-8')
print (df.head(3))
```

```python
def parse_area(s_geocoder):
    a = s_geocoder.split(',')
    res = "(null)"
    try:
        res = a[-6]
    except IndexError as strerror:
        print("IndexError: " + str(strerror))
    except:
        import sys
        print("Unexpected error:", sys.exc_info()[0])
    import re
    res = re.sub("^[ ]*(район )*", "", res)
    print("parse_area")
    #print(s_geocoder)
    #print(res)
    return res

def parse_district(s_geocoder):
    a = s_geocoder.split(',')
    res = "(null)"
    try:
        res = a[-5]
    except IndexError as strerror:
        print("IndexError: " + str(strerror))
    except:
        import sys
        print("Unexpected error:", sys.exc_info()[0])
    import re
    res = re.sub("^[ ]*(округ )*", "", res)
    print("parse_district")
    print(s_geocoder)
    print(res)
    return res

df_res = df
df_res['camera_district'] = df_res.apply(lambda row: parse_district(str(row['camera_geocoder'])), axis=1)
df_res['camera_area'] = df_res.apply(lambda row: parse_area(str(row['camera_geocoder'])), axis=1)
print (df_res.head(3))
```

```python
df_res.to_csv(os.path.join(data_dir, 'ao.csv'), index=False, sep='~', quoting=None, encoding='utf-8')
print("ok")
```
