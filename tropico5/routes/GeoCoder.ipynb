```python
from geopy.geocoders import Nominatim
# READ https://github.com/geopy/geopy
# To find the address corresponding to a set of coordinates:
def ask_geocoder(s_long, s_lat):
    import time
    from random import randint
    time.sleep(randint(0,2))
    res = "-"
    try:
        geolocator = Nominatim()
        location = geolocator.reverse(s_lat + ", " + s_long)
        res = str(location.address)
    except IOError as strerror:
        print("IOError: " + str(strerror))
    except:
        import sys
        print("Unexpected error:", sys.exc_info()[0])
    print(res)
    return res

def parse_district(s_geocoder):
    a = s_geocoder.split(',')
    res = "-"
    try:
        res = a[-6]
    except IndexError as strerror:
        print("IndexError: " + str(strerror))
    except:
        import sys
        print("Unexpected error:", sys.exc_info()[0])
    print(s_geocoder)
    print(res)
    return res

df_res = df_dict
df_res['camera_geocoder'] = df_res.apply(lambda row: ask_geocoder(str(row['camera_longitude']),
                                                                  str(row['camera_latitude'])), axis=1)
df_res['camera_district'] = df_res.apply(lambda row: parse_district(str(row['camera_geocoder'])), axis = 1)

print (df_res.head(3))
```

```python
df_res.to_csv(os.path.join(data_dir, 'result.csv'), index=False, sep=';', quoting=None, encoding='utf-8')
print("ok")
```
