```python
#!pip install folium
WORKDIR="/home/user/ALBINA/RITA/FOLIUM"
OK="ok"
import folium
WHERE = [55.7520,37.6175]
map_osm = folium.Map(location=WHERE, zoom_start=10)
OK
```

```python
import pandas as pd
cams = pd.read_csv(WORKDIR+"/result.csv", encoding='cp1251')
cams = cams[cams['d_camera_tmp_yurbasov.camera_longitude'].notnull()]
cams['geo'] = cams.apply(lambda row: [float(str(row['d_camera_tmp_yurbasov.camera_latitude'])),
                                      float(str(row['d_camera_tmp_yurbasov.camera_longitude']))], axis=1)
cams['codd.v_azimut'] = 1
cams['codd.v_direction'] = cams.apply(lambda row: int(str(row['d_camera_tmp_yurbasov.camera_rank'])) % 2, axis=1)
cams
```

```python
from folium import FeatureGroup
# https://ocefpaf.github.io/python4oceanographers/blog/2015/11/16/folium_quiver/
for i, row in cams.iterrows():
    #print row['geo']
    marker = folium.RegularPolygonMarker(row['geo'], 
                                fill_color='#00aa11' if str(row['codd.v_direction']) == '1' else '#aa1100',
                                number_of_sides=3, 
                                popup=str(row['d_camera_tmp_yurbasov.camera_latitude']) +', '+  str(row['d_camera_tmp_yurbasov.camera_longitude']),
                                radius=15, 
                                rotation=int(row['codd.v_azimut'])-90)
    feature_group = folium.FeatureGroup(name='#'+str(i))
    feature_group.add_child(marker)
    map_osm.add_child(marker)
```

```python
map_osm.save(WORKDIR + '/Stable.html')
OK
```
